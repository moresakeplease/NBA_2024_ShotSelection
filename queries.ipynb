<h1>Finding Shot Percentage</h1>

```python
shots_made_query = "SELECT COUNT(SHOT_MADE) AS made FROM shots WHERE PLAYER_NAME = 'Luka Doncic' and SHOT_MADE = true"
shots_total_query = "SELECT COUNT(SHOT_MADE) As total FROM shots WHERE PLAYER_NAME = 'Luka Doncic'"

shots_total_results = pd.read_sql_query(shots_total_query, con)
shots_made_results = pd.read_sql_query(shots_made_query, con)

shots_made = shots_made_results['made'].iloc[0]
shots_total = shots_total_results['total'].iloc[0]

print('shots_total:', shots_total)
print('shots_made:', shots_made)
```

```python
if shots_total > 0:
    shot_percentage = shots_made / shots_total
else:
    print('Luka didnt take shot')

print('shot_percentage:', shot_percentage)
```

<h1>Grouping by Zone</h1>

```python
shot_zone_query = "SELECT BASIC_ZONE, ZONE_NAME, ZONE_RANGE, COUNT(*) AS SHOT_COUNT, \
SUM(CASE WHEN SHOT_MADE = true THEN 1 ELSE 0 END) AS SHOT_MADE, \
SUM(CASE WHEN SHOT_MADE = false THEN 1 ELSE 0 END) AS SHOT_MISSED, \
CASE WHEN COUNT(*) = 0 THEN 0 ELSE (SUM(CASE WHEN SHOT_MADE = true THEN 1 ELSE 0 END) * 100.0 / COUNT(*)) END AS SHOT_PERCENTAGE \
FROM shots WHERE PLAYER_NAME = 'Luka Doncic' GROUP BY BASIC_ZONE"

shot_zone_results = pd.read_sql_query(shot_zone_query, con)
shot_zone_results
```

<h1>Luka VS Opposing Team</h1>

```python
shot_versus_query = "SELECT HOME_TEAM, AWAY_TEAM, GAME_DATE, COUNT(*) AS SHOT_COUNT, \
SUM(CASE WHEN SHOT_MADE = true THEN 1 ELSE 0 END) AS SHOT_MADE, \
SUM(CASE WHEN SHOT_MADE = false THEN 1 ELSE 0 END) AS SHOT_MISSED, \
CASE WHEN COUNT(*) = 0 THEN 0 ELSE (SUM(CASE WHEN SHOT_MADE = true THEN 1 ELSE 0 END) * 100.0 / COUNT(*)) END AS SHOT_PERCENTAGE \
FROM shots WHERE PLAYER_NAME = 'Luka Doncic' GROUP BY GAME_DATE ORDER BY HOME_TEAM DESC"

shot_versus_results = pd.read_sql_query(shot_versus_query, con)
shot_versus_results
```

<h1>Luka Favorite Action Type</h1>

```python
shot_action_query = "SELECT ACTION_TYPE, COUNT(*) AS SHOT_COUNT, \
SUM(CASE WHEN SHOT_MADE = true THEN 1 ELSE 0 END) AS SHOT_MADE, \
SUM(CASE WHEN SHOT_MADE = false THEN 1 ELSE 0 END) AS SHOT_MISSED, \
CASE WHEN COUNT(*) = 0 THEN 0 ELSE (SUM(CASE WHEN SHOT_MADE = true THEN 1 ELSE 0 END) * 100.0 / COUNT(*)) END AS SHOT_PERCENTAGE \
FROM shots WHERE PLAYER_NAME = 'Luka Doncic' GROUP BY ACTION_TYPE ORDER BY SHOT_COUNT DESC"

shot_action_results = pd.read_sql_query(shot_action_query, con)
shot_action_results
```

<h1>Best Step Back Shooter</h1>

```python
step_back_query = "SELECT ACTION_TYPE, PLAYER_NAME, COUNT(*) AS SHOT_COUNT, \
SUM(CASE WHEN SHOT_MADE = true THEN 1 ELSE 0 END) AS SHOT_MADE, \
SUM(CASE WHEN SHOT_MADE = false THEN 1 ELSE 0 END) AS SHOT_MISSED, \
CASE WHEN COUNT(*) = 0 THEN 0 ELSE (SUM(CASE WHEN SHOT_MADE = true THEN 1 ELSE 0 END) * 100.0 / COUNT(*)) END AS SHOT_PERCENTAGE \
FROM shots WHERE ACTION_TYPE = 'Step Back Jump shot' GROUP BY PLAYER_NAME ORDER BY SHOT_COUNT DESC"

step_back_result = pd.read_sql_query(step_back_query, con)
step_back_result.head(10)
```
